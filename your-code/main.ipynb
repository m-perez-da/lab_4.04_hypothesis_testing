# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here
# Two-tailed test
# Null hypothesis H0: Hourly wage of hourly workers is $30/hr
# Alternative hypothesis H1: Hourly wage of hourly workers is significantly different from $30/hr
hourly_workers=salaries[salaries['Salary or Hourly']=='Hourly']
hourly_workers['Hourly Rate']=list(map(lambda x: x.replace('$', ''), hourly_workers['Hourly Rate']))
hourly_workers['Hourly Rate']=list(map(float, hourly_workers['Hourly Rate']))
hourly_rates=hourly_workers['Hourly Rate']

import scipy.stats

confidence_level = 0.95
degrees_freedom = len(hourly_rates) - 1
sample_mean = np.mean(hourly_rates)
sample_standard_error = scipy.stats.sem(hourly_rates)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
print( 'confidence interval is ', confidence_interval, '.' )

from scipy.stats import ttest_1samp
print('stat is  ', stat)
stat, pval = ttest_1samp(hourly_rates, 30)
print('pvalue for the two-tailed test is ', pval)
# Since p-value is very low (<0.05) we reject H0 (hourly wage is $30/hr)
```

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our hypothesis is greater than, then to reject, the test statistic must also be positive.

```python
# Your code here:
# One-tailed test
# Null hypothesis H0: Salaries of department of police are $86000/year
# Alternative hypothesis H1: Salaries of department of police are higher than $86000/year
police_salaries=salaries[((salaries['Department']=='POLICE') & (salaries['Salary or Hourly']=='Salary'))]
police_salaries['Annual Salary']=list(map(lambda x: x.replace('$', ''), police_salaries['Annual Salary']))
police_salaries['Annual Salary']=list(map(float, police_salaries['Annual Salary']))
salary_rates=police_salaries['Annual Salary']

confidence_level = 0.95
degrees_freedom = len(salary_rates) - 1
sample_mean = np.mean(salary_rates)
sample_standard_error = scipy.stats.sem(salary_rates)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
print( 'confidence interval is ', confidence_interval, '.' )

from scipy.stats import ttest_1samp
print('stat is  ', stat)
stat, pval = ttest_1samp(salary_rates, 86000)
print('pvalue for the one-tailed test is ', pval/2)
# Since p-value is very low (<0.05) we reject H0 (mean salary is $86000/year)
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here:
hourly_workers=salaries[salaries['Salary or Hourly']=='Hourly']
hourly_workers['Hourly Rate']=list(map(lambda x: x.replace('$', ''), hourly_workers['Hourly Rate']))
hourly_workers['Hourly Rate']=list(map(float, hourly_workers['Hourly Rate']))
hourly_rates=hourly_workers['Hourly Rate']
confidence_level = 0.95
degrees_freedom = len(hourly_rates) - 1
sample_mean = np.mean(hourly_rates)
sample_standard_error = scipy.stats.sem(hourly_rates)

confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
print( 'confidence interval is ', confidence_interval, '.' )
```

# Bonus Challenge - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
# Your code here:
# Null hypothesis H0: p= 25
# Alternative Hypothesis H1: p 25
from statsmodels.stats.proportion import proportions_ztest
stat, pval = proportions_ztest(count=len(salaries[salaries['Salary or Hourly']=='Hourly']), nobs=len(salaries), value=0.25, alternative='two-sided')
stat, pval
```

```python
# Since p-value=0.00027 is less than  = 0.05, we reject the null hypothesis, and we conclude the number of hourly workers in the City 
# of Chicago is significantly different from 25%
```
